```python
# Remove spaces from column names
df.columns = df.columns.str.strip()
df.columns
```

```python
# Convert delivery time into numbers
df['Time_taken(min)'] = df['Time_taken(min)'].str.strip().str.extract(r'(\d+)')
df['Time_taken(min)'] = df['Time_taken(min)'].astype(int)
df['Time_taken(min)'].head()
```

```python
# Convert order date to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y', dayfirst=True)
df.info()
```

```python
text_columns = [
    'Weatherconditions',
    'Road_traffic_density',
    'Type_of_order',
    'Type_of_vehicle',
    'Festival',
    'City'
]
for col in text_columns:
    df[col] = df[col].astype(str).str.strip()
print('Text cleaning completed')
```

```python
df.info()
```

```python
from geopy.distance import geodesic

restaurant_point = (df.loc[0, 'Restaurant_latitude'], df.loc[0, 'Restaurant_longitude'])
delivery_point = (df.loc[0, 'Delivery_location_latitude'], df.loc[0, 'Delivery_location_longitude'])
distance_km = geodesic(restaurant_point, delivery_point).kilometers
print(f'Row 0 distance: {distance_km:.2f} km')

def compute_distance_km(row):
    restaurant_point = (row['Restaurant_latitude'], row['Restaurant_longitude'])
    delivery_point = (row['Delivery_location_latitude'], row['Delivery_location_longitude'])
    return geodesic(restaurant_point, delivery_point).kilometers
```
